Introduction to Object Oriented Method

By using Object Ori3ented method, we can create figure objects and then just call methods or attributes of that object. This approach is nicer when dealing with a canvas that has multiple plots on it.

```python
x = np.linspace(0, 5, 11)
y=x**2
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y,'b')
axes.set_xlabel('X axis')
axes.set_ylabel('Y axis')
axes.set_title('Plotting data with matplotlib')
plt.show();
```

Inset Graphs

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8]) # main axes
axes2 = fig.add_axes([0.175,0.5,0.4,0.3]) # inset axes

# main figure
axes1.plot(x,y,'b-->')
axes1.set_xlabel('X axis')
axes1.set_ylabel('Y axis')
axes1.set_title('Main Plot')

# inset graph
axes2.plot(y,x,'r<--')
axes2.set_xlabel('X axis')
axes2.set_ylabel('Y axis')
axes2.set_title('Inset Plot');
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8]) # main axes
axes2 = fig.add_axes([0.5,0.8,0.4,0.3]) # inset axes

# main figure
axes1.plot(x,y,'b-->')
axes1.set_xlabel('X axis')
axes1.set_ylabel('Y axis')
axes1.set_title('Main Plot')

# inset graph
axes2.plot(y,x,'r<--')
axes2.set_xlabel('X axis')
axes2.set_ylabel('Y axis')
axes2.set_title('Inset Plot');
```

```python
fig, axes = plt.subplots()

axes.plot(x,y,'b*')
axes.set_xlabel('X axis')
axes.set_ylabel('Y axis')
axes.set_title('Plotting data with matplotlib - subplots');
```

```python
figs, axes = plt.subplots(nrows=1, ncols=4)
```

```python
axes
```

```python
for ax in axes:
    ax.plot(x,y,'b')
    ax.set_xlabel('X axis')
    ax.set_ylabel('Y axis')
    ax.set_title('subplots 2')

# Display the figure object
figs
```
