### Basic Matplotlib commands


1. `plot()`: This command is used to plot a line or scatter plot in `matplotlib`. It takes in two arrays as arguments: the x-axis values and the y-axis values. For example, `plot(x_values, y_values)` would plot a line or scatter plot with the x-axis values given by the `x_values` array and the y-axis values given by the `y_values` array.

2. `xlabel()`: This command is used to set the label for the x-axis of the plot. For example, `xlabel('Time (s)')` would set the x-axis label to "Time (s)".

3. `ylabel()`: This command is used to set the label for the y-axis of the plot. For example, `ylabel('Temperature (C)')` would set the y-axis label to "Temperature (C)".

4. `title()`: This command is used to set the title of the plot. For example, `title('Temperature vs. Time')` would set the title of the plot to "Temperature vs. Time".

5. `show()`: This command is used to display the plot on the screen. It is usually the last command in a `matplotlib` script. For example, `plot(x_values, y_values); xlabel('x-axis'); ylabel('y-axis'); title('My Plot'); show()` would create a plot with x-axis labeled "x-axis", y-axis labeled "y-axis", and title "My Plot" and display it on the screen.

````python
plt.plot(x,y,'r')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Plotting data with matplotlib')
# plt.grid(axis='x')
plt.grid()
plt.show()
````

### Creating multiplots on the same canvas

`subplot()`
This command is used to create multiple plots on the same canvas. It takes in three arguments: the number of rows, the number of columns, and the index of the plot to create. For example, `subplot(2, 1, 1)` would create a plot in the top row of a 2x1 grid, while `subplot(2, 1, 2)` would create a plot in the bottom row of the same grid.
`grid()`
This command is used to display a grid on the plot. For example, `grid()` would display a grid on the plot. `grid(True)` would display a grid on the plot, while `grid(False)` would remove the grid from the plot.

### Line style

The line style of a graph is the pattern of dashes and dots that is used to draw the lines in the graph. In `matplotlib`, the line style of a graph can be set using the `linestyle` parameter of the `plot()` function. Here are some examples of different line styles:
 - Solid line: `plot(x_values, y_values, linestyle='solid')`
 - Dashed line: `plot(x_values, y_values, linestyle='dashed')`
 - Dotted line: `plot(x_values, y_values, linestyle='dotted')`
 - Dash-dot line: `plot(x_values, y_values, linestyle='dashdot')`
 - Dash-dot-dot line: `plot(x_values, y_values, linestyle='dashdotdot')`
 - No line: `plot(x_values, y_values, linestyle='none')`
 - Custom line style: `plot(x_values, y_values, linestyle=(0, (1, 1)))`
You can use the short hand as follows:
 - Solid line: `plot(x_values, y_values, '-')`
 - Dashed line: `plot(x_values, y_values, '--')`
 - Dotted line: `plot(x_values, y_values, ':')`
 - Dash-dot line: `plot(x_values, y_values, '-.')`
 - Dash-dot-dot line: `plot(x_values, y_values, '-..')`
 - No line: `plot(x_values, y_values, 'none')`

### Introduction to Object Oriented Method

In `matplotlib`, object-oriented programming (OOP) is an alternative approach to creating plots that provides more control and flexibility over the plot creation process. 

Matplotlib, like many other Python libraries, is designed using object-oriented programming (OOP) principles. This means that you can create objects and manipulate them using methods and attributes. Understanding the OOP structure of Matplotlib can help you to create more complex and customized plots.

In Matplotlib, the main object is the `Figure`, which represents the entire figure that you are plotting. A `Figure` object can contain one or more subplots, represented by `Axes` objects. Each `Axes` object represents a single plot, and has its own set of attributes and methods for customizing its appearance and behavior.

To create a simple plot using the object-oriented method, you first create a `Figure` object, and then add one or more `Axes` objects to it. You can then use the methods and attributes of the `Axes` objects to customize the plot.

Using the object-oriented method, you have more control over the appearance and behavior of your plots, and can create more complex plots with multiple subplots and customized layouts.

### Inset Graphs

An inset graph is a small graph that is placed inside a larger graph. It is used to show a zoomed-in view of a particular region of the larger graph. In `matplotlib`, inset graphs can be created using the `inset_axes()` function. 

````python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8]) # main axes
axes2 = fig.add_axes([0.175,0.5,0.4,0.3]) # inset axes

# main figure
axes1.plot(x,y,'b-->')
axes1.set_xlabel('X axis')
axes1.set_ylabel('Y axis')
axes1.set_title('Main Plot')

# inset graph
axes2.plot(y,x,'r<--')
axes2.set_xlabel('X axis')
axes2.set_ylabel('Y axis')
axes2.set_title('Inset Plot');
````

### Aspect Ratio, DPI and Figure Size

1. Aspect Ratio: Aspect ratio refers to the proportion of the width and height of your figure. It is defined as the ratio of the height to the width of the figure. The aspect ratio can affect the perception of your data, as it can stretch or compress the data in one direction. You can set the aspect ratio of your figure using the `aspect` parameter of the `imshow()` or `plot()` functions.

2. DPI (dots per inch): DPI refers to the number of dots (or pixels) per inch in your figure. It affects the resolution and quality of your figure. The higher the DPI, the more detailed and clear your figure will appear. You can set the DPI of your figure using the `dpi` parameter of the `savefig()` function.

3. Figure Size: Figure size refers to the physical dimensions of your figure in inches. It affects how much space your figure occupies on the screen or in print. You can set the figure size using the `figsize` parameter of the `subplots()` or `figure()` functions.


````python
x = np.linspace(0, 5, 11)
y = x ** 2
````

````python
fig = plt.figure(figsize=(8,4), dpi=100)

# Figure size x * dpi = width, y * dpi = height
````

````python
fig, axes = plt.subplots(figsize=(4,3), dpi=100)

axes.plot(x,y,'r')
axes.set_xlabel('X axis')
axes.set_ylabel('Y axis')
axes.set_title('Plotting data with matplotlib - Modifying size and DPI');
````

### Legends

You can use the `label` argument in the `plot()` function to add a label to a line or scatter plot. For example, `plot(x_values, y_values, label='My Line')` would add the label "My Line" to the line or scatter plot.

````python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x, x**2, label="X Squared")
ax.plot(x, x**3, label="X Cubed")
ax.legend()
ax.set_title("Plotting data with matplotlib - Legend");
````

### How loc in legend works

You can use the `legend()` function to add a legend to a plot. For example, `legend(loc='upper left')` would add a legend to the plot with the legend labels in the upper left corner of the plot. The `loc` argument can take on several different values, including `upper right`, `upper left`, `lower right`, `lower left`, `right`, `center left`, `center right`, `lower center`, `upper center`, and `center`. The default value for `loc` is `best`, which automatically chooses the best location for the legend. You can also pass number to the `loc` argument to specify the exact location of the legend. For example, `legend(loc=2)` would add a legend to the plot with the legend labels in the upper left corner of the plot.

````python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x, x, label="X")
ax.plot(x, x**4, label="X Raised to 4")
ax.legend(loc=10)
ax.set_title("Plotting data with matplotlib - Legend");
````

The `marker` argument can take on several different values, including `o`, `s`, `^`, `*`, `+`, `x`, `D`, `d`, `h`, `H`, `p`, `v`, `<`, `>`, and `.`. The default value for `marker` is `None`, which plots a line without markers. You can also pass a tuple to the `marker` argument to specify the exact style of the markers. For example, `plot(x_values, y_values, marker=('o', 's'))` would plot a line or scatter plot with circular markers and square markers.

### Params in `plot()`
In Matplotlib, plot parameters are used to customize various aspects of a plot such as the axis labels, title, color, line style, marker style, and many more. These parameters can be adjusted using various methods provided by the Matplotlib library.

Here are some common plot parameters that you can adjust in Matplotlib:

1. `xlabel`, `ylabel`: These parameters are used to set the labels for the x and y-axes, respectively.

2. `title`: This parameter is used to set the title of the plot.

3. `xlim`, `ylim`: These parameters are used to set the limits of the x and y-axes, respectively.

4. `grid`: This parameter is used to enable or disable the grid lines on the plot.

5. `color`: This parameter is used to set the color of the plot elements such as lines, markers, and text.

6. `linestyle`: This parameter is used to set the style of the plot lines, such as solid, dashed, or dotted.

7. `marker`: This parameter is used to set the style of the plot markers, such as circles, squares, or triangles.

8. `markersize`: This parameter is used to set the size of the plot markers.

9. `linewidth`: This parameter is used to set the width of the plot lines.


````python
fig, ax = plt.subplots()

ax.plot(x, x+1, color="blue", alpha=0.5) # half-transparant
ax.plot(x, x+2, color="#8B008B")        # RGB hex code
ax.plot(x, x+3, color="#FF8C00")        # RGB hex code
ax.set_title("Plotting data with matplotlib - Colors");
````

````python
fig, ax = plt.subplots(figsize=(12,6))

ax.plot(x, x+1, color="cyan", lw=1, ls='-', marker='o', markersize=5)
ax.plot(x, x+2, color="pink", lw=1, ls='-', marker='s', markersize=10)
ax.plot(x, x+3, color="grey", lw=1, ls='--', marker='^', markersize=20, markerfacecolor='white', markeredgewidth=5, markeredgecolor='pink')
ax.plot(x, x+4, color="pink", lw=1, ls='-', marker='*', markersize=10)
ax.plot(x, x+5, color="cyan", lw=1, ls='-', marker='<', markersize=5)
ax.set_title("Plotting data with matplotlib - Markers");
````

### Control over axis appearance

You can use the `axis()` function to control the appearance of the x-axis and y-axis. For example, `axis([0, 10, 0, 10])` would set the x-axis limits to 0 and 10 and the y-axis limits to 0 and 10. The `axis()` function can take on several different values, including `off`, `equal`, `scaled`, `tight`, `auto`, `image`, `square`, `tight`, `nor` and `scaled`. The default value for `axis` is `auto`, which automatically chooses the axis limits based on the data. You can also pass a tuple to the `axis` argument to specify the exact limits of the x-axis and y-axis. For example, `axis([0, 10, 0, 10])` would set the x-axis limits to 0 and 10 and the y-axis limits to 0 and 10.

Plot Range

You can use the `xlim()` and `ylim()` functions to control the range of values displayed on the x-axis and y-axis. For example, `xlim(0, 10)` would set the x-axis limits to 0 and 10. The `xlim()` and `ylim()` functions can take on several different values, including `auto`, `tight`, `scaled`, `equal`, `off`, `image`, `square`, `tight`, `nor` and `scaled`. The default value for `xlim` and `ylim` is `auto`, which automatically chooses the axis limits based on the data. You can also pass a tuple to the `xlim` and `ylim` arguments to specify the exact limits of the x-axis and y-axis. For example, `xlim(0, 10)` would set the x-axis limits to 0 and 10.

````python
fig, axes = plt.subplots(1,3, figsize=(12,4))

axes[0].plot(x, x**2, x, x**3)
axes[0].set_title("default axes ranges")

axes[1].plot(x, x**2, x, x**3)
axes[1].axis('tight')
axes[1].set_title("tight axes")

axes[2].plot(x, x**2, x, x**3)
axes[2].set_ylim([0, 60])
axes[2].set_xlim([2, 5])
axes[2].set_title("custom axes range");
````
