```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

Basic Matplotlib commands

```python
plt.plot(x,y,'r')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Plotting data with matplotlib')
# plt.grid(axis='x')
plt.grid()
plt.show()
```

Creating multiplots on the same canvas

```python
plt.subplot(1,3,1)
plt.plot(x,y,'r--') 
plt.subplot(1,3,2)
plt.grid(color='blue',linestyle='--', linewidth=0.5)
plt.plot(y,x,'g*--')
plt.subplot(1,3,3)
plt.plot(y,y,'m.-.');
```

```python
plt.scatter(x,y);
```
